```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
 
df = pd.read_excel('word_list.xlsx', sheetname='Sheet1')
 
print("Column headings:")
print(df.columns)
```

```python
word_list_dirty = df['Word'].astype('str')
word_list =[]
for word in word_list_dirty:
    word_list.append(word[3:])
word_list
```

```python
import bs4
import requests
sess = requests.Session()
adapter = requests.adapters.HTTPAdapter(max_retries = 20)
sess.mount('http://', adapter)
def get_soup(word):
    url = 'https://www.merriam-webster.com/dictionary/%s' %word
    response = sess.get(url)
    html = response.text
    soup = bs4.BeautifulSoup(html, "html.parser")
    return soup
```

```python
import re
def get_word_class(soup):
    mydivs = soup.findAll("div", {"class": "entry-attr"})
    lines = [span.get_text() for span in mydivs]
    if len(lines)>0:
        if len(lines[0])>2: 
            defini_line = lines[0][2:20]
    else:
        defini_line = 'NA'
    good_line = defini_line.replace('\n','')
    return good_line
```

```python
i = 1
word_class_dict = {}
word_set= set(word_list)
word_list = list(word_set)
for word in word_list[:1000]:
    soup = get_soup(word)
    word_class = get_word_class(soup)
    print(word)
    if 'verb' in word_class and 'adverb' not in word_class:
        word_class_dict[word] = 'verb'
        print("     verb", " ", len(word_class_dict))
    elif 'conjunction' in word_class:
        word_class_dict[word] = 'conjunction'
        print("     conjunction", " ", len(word_class_dict))
    elif 'preposition' in word_class:
        word_class_dict[word] = 'preposition'
        print("     preposition", " ", len(word_class_dict))
    elif 'noun' in word_class and 'pronoun' not in word_class:
        word_class_dict[word] = 'noun'
        print("     noun", " ", len(word_class_dict))
    elif 'pronoun' in word_class:
        word_class_dict[word] = 'pronoun'
        print("     pronoun", " ", len(word_class_dict))
    elif 'adjective' in word_class:
        word_class_dict[word] = 'adjective'
        print("     adjective", " ", len(word_class_dict))
    elif 'adverb' in word_class:
        word_class_dict[word] = 'adverb'
        print("     adverb", " ", len(word_class_dict))
    elif 'abbreviation' in word_class:
        word_class_dict[word] = 'abbreviation'
        print("     abbreviation", " ", len(word_class_dict))
    elif 'definite article' in word_class:
        word_class_dict[word] = 'definite_article'
        print("     definite_article", " ", len(word_class_dict))
    elif  word == "these" or word == "those":
        word_class_dict[word] = 'pronoun'
        print("     pronoun", " ", len(word_class_dict))
    else:
        word_class_dict[word] = 'NA'
        print("     NA", " ", len(word_class_dict))
```
